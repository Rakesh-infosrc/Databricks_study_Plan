# Databricks Serverless SQL Warehouses

Serverless SQL Warehouses provide an elastic, managed environment for SQL workloads, optimizing both performance and cost.

## Key Advantages

- **Instant & Elastic Compute:** Instantly available environment that scales dynamically, eliminating infrastructure wait times and over-provisioning costs
- **Reduced Management Overhead:** Databricks handles capacity management, patching, upgrades, and performance optimization
- **Predictable Pricing:** Simplified, usage-based model with no idle time charges — system auto-shuts-down or scales based on workload

## Core Performance Features

- **Photon:** Native vectorized query engine that accelerates SQL and DataFrame API calls
- **Predictive I/O:** Uses machine learning to accelerate read/write/update operations and efficient point lookups, without the cost of maintaining traditional search indexes
- **Intelligent Workload Management:** Uses machine learning to route queries, prioritize tasks, and preemptively scale the cluster to maximize throughput and minimize latency

## Configuring a SQL Warehouse

Settings configured in the Databricks UI:

- **Cluster Size:** Sets the driver node and worker count (default is extra large)
- **Auto Stop:** Minutes of inactivity before the warehouse stops, to save on DBUs and cloud instance charges
- **Scaling:** Defines the min/max number of clusters to handle concurrent query demand
- **Type:** "Serverless" is recommended for the fastest startup, efficient I/O, and rapid auto-scaling
