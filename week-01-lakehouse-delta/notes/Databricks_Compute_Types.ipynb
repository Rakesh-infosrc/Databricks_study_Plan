# Databricks Compute Types — Overview

## 1. All-Purpose Compute
**Use Case:** Development, exploration, notebooks, ad-hoc analysis.

**Features:**
- Supports multiple users working together
- Used with notebooks, dashboards, and APIs
- Costs more — always running while in use
- Good for data exploration and testing pipelines

## 2. Job Compute
**Use Case:** Running production jobs or scheduled pipelines.

**Features:**
- Spins up when a job starts, terminates when the job ends
- More cost-effective than all-purpose clusters
- Can't be used for interactive notebooks

## 3. Compute Pools
**Use Case:** Reduce startup time and cost for clusters.

**Features:**
- Pre-warmed compute resources
- Saves cost by avoiding cold-start times
- Usable for both job and all-purpose clusters

## 4. SQL Warehouse
**Use Case:** Running SQL queries and powering dashboards.

**Features:**
- Connects easily to BI tools like Power BI, Tableau
- Optimised for SQL queries, not Spark jobs

## 5. Serverless Compute
**Use Case:** Hands-off resource management — best for cost and scaling.

**Features:**
- Auto-scales and auto-terminates
- Databricks manages instances, not you
- Ideal for unpredictable workloads
